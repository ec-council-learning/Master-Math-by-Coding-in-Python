```python
# import libraries
import numpy as np
```

```python
aList = [0,1,2,3,4,5]
aList
```

```python
type(aList)
```

```python
listList = [ 3,['3','4','5'],5,[4,5,6] ]
listList
```

```python
listList = [
            3,              # a number
            ['3','h','5'],  # a list of strings
            5,              # another number
            [4,5,6]         # a list of numbers
           ]

listList
```

```python
# can we add lists?

aList = [0,1,2,3,4,5]
strList = ['0','1','2','3','4','5']

aList + strList
# aList + listList
```

```python
# multiply a list?
aList*3

# the principal here is 'overloading': One operation has multiple behaviors depending on the context
```

```python
# lists vs. numpy arrays

aList_np = np.array(aList)

print(aList)
print(aList_np)
```

```python
# now let's try multiplication again
print(aList*3)
print(aList_np*3)
```

```python
# scalar addition?
print(aList+3)
print(aList_np+3)
```

```python
# regular method
for idx,item in enumerate(listList):
  print(f'List element {idx} is {item} and is of type {type(item)}.')
```

```python
# advanced method to print out only the type
for idx,item in enumerate(listList):
  t = str(type(item))[8:]
  apost = t.find("'")
  print(f'List element {idx} is {item} and is of type {t[:apost]}.')
```
