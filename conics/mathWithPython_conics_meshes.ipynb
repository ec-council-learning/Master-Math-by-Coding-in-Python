```python
# import libraries
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
X,Y = np.meshgrid(range(0,10),range(0,15))

plt.subplot(121)
plt.pcolormesh(X,edgecolors='k',linewidth=.1)
# plt.gca().set_aspect('equal')
plt.title('X')

plt.subplot(122)
plt.pcolormesh(Y,edgecolors='k',linewidth=.1)
plt.gca().set_aspect('equal')
plt.title('Y')

plt.show()
```

```python
x = np.linspace(0,2*np.pi,50)
y = np.linspace(0,4*np.pi,50)

X,Y = np.meshgrid(x,y)

F = np.cos(X) + np.sin(Y)
plt.imshow(F,extent=[x[0],x[-1],y[0],y[-1]])
plt.show()
```

```python
# Make a gaussian
x = np.linspace(-2,2,100)
s = 2

# create the gaussian
X,Y = np.meshgrid(x,x)
gauss2d = np.exp( -(X**2+Y**2)/s )

# and plot
plt.imshow(gauss2d,extent=[x[0],x[-1],x[0],x[-1]])
plt.axis('off')
plt.show()
```
