```python
# import libraries
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
# parameters
n = 100
a = 1
h = 1
k = -2

# x-axis points to evaluate the function
x = np.linspace(-2,4,n)

# create the function
y = a*(x-h)**2 + k

# and plot it!
plt.plot(x,y)
plt.grid()
plt.axis('square')
plt.show()
```

```python
# draw the vertex, focus, and directrix

x = np.linspace(-6,6,40)
y = (x-2)**2/2 + 1

# 1) convert to standard parabola equation
# 4*p*(y-k) = (x-h)**2
# 2) write standard equation
h = 2
k = 1
p = 1/2 # 2=4p


# plot the parabola
plt.plot(x,y,label='Parabola')

# plot the vertex
plt.plot(h,k,'ro',label='Vertex')

# plot the focus
plt.plot(h,k+p,'go',label='focus')

# plot the directrix
d = k-p
plt.plot(x[[0,-1]],[d,d],label='directrix')

plt.legend()
plt.axis('square')
plt.axis([x[0],x[-1],d-.5,10])
plt.grid()
plt.show()
```
