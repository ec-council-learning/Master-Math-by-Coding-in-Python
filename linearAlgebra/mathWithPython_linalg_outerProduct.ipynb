```python
v1 = np.random.randn(50)
v2 = np.random.randn(80)

np.dot(v1,v2)
```

# Exercise

```python
v = np.arange(1,10)
w = np.arange(1,5)

print(np.outer(v,w))
print(np.outer(w,v))
```

```python
# show that svw=vsw=vws
s = 4

left = s*np.outer(v,w)
middle = np.outer(v*s,w)
right = np.outer(v,w)*s

print(left-middle)
print(middle-right)
print(left-right)
```
