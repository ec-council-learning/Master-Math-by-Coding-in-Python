```python
# report the absolute value of a set of numbers only if they are less than -5 or greater than +2
numbers = [-4,6,-1,43,-18,2,0]

# for-loop
for i in numbers:
  if i<-5 or i>2:
    print( 'Absolute value of %g is %g.' %(i,abs(i)) )
  else:
    print( str(i) + ' was not tested.' )
```
