```python
x = 5
y = 5.1

# 1)
ans = x**(3/4) * 4**y
display(Math('x^{3/4}\\times 4^y = %g'%ans))

# 2)
solution = (3**3) / (x**y)
display(Math('\\frac{3^3}{x^y} = %g'%solution))

# 3)
ans = 10**(x-4)
display(Math('10^{x-4} = %g'%ans))
```
