```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,1,2001)

y = ( 1 - x**4 )**(1/4)

plt.plot(x,y,'k',linewidth=3)
plt.plot(x,-y,'k',linewidth=3)

plt.axis('square')
plt.axis('off')
plt.show()
```
