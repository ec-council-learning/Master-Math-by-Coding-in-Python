```python
import sympy as sym
from IPython.display import display,Math
```

```python
mu,alpha = sym.symbols('mu,alpha')

expr = 2*sym.exp(mu**2/alpha)

display(Math( sym.latex(expr) ))
```

```python
x = sym.symbols('x')
sym.solve( 4*x - 2 )
```

```python
# part 1 of 2

q = x**2
r = x**2

display(q)
display(r)
```

```python
# part 2 of 2

q,r = sym.symbols('q,r')

q = sym.sympify('x^2')
r = sym.sympify('x**2')

display(Math(sym.latex(q)))
display(r)

sym.Eq(q,r)
```
