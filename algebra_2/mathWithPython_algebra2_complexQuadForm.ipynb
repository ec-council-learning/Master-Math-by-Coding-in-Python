```python
# import libraries and functions
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

from numpy.lib.scimath import sqrt
```

```python
# as a function!

def quadeq(a,b,c):

  # initialize
  out = np.zeros(2,dtype=complex)

  # compute
  out[0] = ( -b - sqrt(b**2 - 4*a*c) ) / (2*a)
  out[1] = ( -b + sqrt(b**2 - 4*a*c) ) / (2*a)

  # output
  return out
```

```python
x = 1
y = range(-5,6)
z = range(-2,11)

M = np.zeros((len(y),len(z)),dtype=complex)

for i in range(0,len(y)):
  for j in range(0,len(z)):
    M[i,j] = quadeq(1,y[i],z[j])[0]


plt.subplot(1,3,1)
plt.imshow(np.real(M))
plt.axis('off'), plt.title('real part')

plt.subplot(132)
plt.imshow(np.imag(M))
plt.axis('off'), plt.title('imaginary part')

plt.subplot(133)
plt.imshow(np.abs(M))
plt.axis('off'), plt.title('magnitude')

plt.show()
```
