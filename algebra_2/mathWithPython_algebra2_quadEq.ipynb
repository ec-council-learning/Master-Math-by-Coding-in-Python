```python
# import libraries and functions
import sympy as sym
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
# the quadratic equation

a = 2
b = 7
c = 5

quadeqP = ( -b + np.sqrt(b**2 - 4*a*c) ) / (2*a)
quadeqM = ( -b - np.sqrt(b**2 - 4*a*c) ) / (2*a)

print(quadeqP,quadeqM)
```

```python
# but use a different sqrt function for complex results...

a = 3
b = 7
c = 5

quadeqP = ( -b + np.lib.scimath.sqrt(b**2 - 4*a*c) ) / (2*a)
quadeqM = ( -b - np.lib.scimath.sqrt(b**2 - 4*a*c) ) / (2*a)

print(quadeqP,quadeqM)
```

```python
# ... or explicitly tell numpy to allow complex-valued solutions

a = 3
b = 7
c = 5

quadeqP = ( -b + np.sqrt(b**2 - 4*a*c,dtype=complex) ) / (2*a)
quadeqM = ( -b - np.sqrt(b**2 - 4*a*c,dtype=complex) ) / (2*a)

print(quadeqP,quadeqM)
```

```python
# observe:
print( np.sqrt(-1) )
print( sp.sqrt(-1) )
print( np.lib.scimath.sqrt(-1) )

# can also just import the function for convenience
from numpy.lib.scimath import sqrt
print( sqrt(-1) )
```

```python
# as a function!

def quadeq(a,b,c):

  # initialize
  out = np.zeros(2,dtype=complex)

  # compute
  out[0] = ( -b - sqrt(b**2 - 4*a*c) ) / (2*a)
  out[1] = ( -b + sqrt(b**2 - 4*a*c) ) / (2*a)

  # output
  return out
```

```python
# test!
print(quadeq(2,7,2))

print(quadeq(1,4,-2))
```

```python
a = 1
b = range(-5,6)
c = range(-2,11)

M = np.zeros((len(b),len(c)),dtype=complex)

for i in range(0,len(b)):
  for j in range(0,len(c)):
    M[i,j] = quadeq(a,b[i],c[j])[0]


plt.imshow(np.abs(M),extent=[ c[0],c[-1],b[0],b[-1] ],origin='lower')
plt.xlabel('c')
plt.ylabel('b')
plt.title('a='+str(a))
plt.colorbar()
plt.show()
```
