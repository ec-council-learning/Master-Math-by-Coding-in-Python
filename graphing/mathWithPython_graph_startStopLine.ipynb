```python
# import libraries
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
# drawing lines

p1 = [-3,-1]
p2 = [4,4]

# nice try, but wrong code :(
plt.plot(p1,p2)
plt.plot([p1[0],p2[0]],[p1[1],p2[1]],color=[.7,.3,.8],linewidth=5)

plt.axis('square')
plt.axis([-6,6,-6,6])
plt.show()
```

```python
x = 3
y = 5

# basic plotting a red dot
plt.plot(x,y,'ro')
plt.plot([0,x],[0,y],'r')

# set axis limits
plt.axis('square') # order matters
plt.axis([-6,6,-6,6])
plt.grid()

# now add lines
plt.plot([-6,6],[0,0],'k')
plt.plot([0,0],[-6,6],'k')

plt.show()
```
