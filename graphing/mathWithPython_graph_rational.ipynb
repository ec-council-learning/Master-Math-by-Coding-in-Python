```python
# import libraries
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0,len(x)):
  y[i] = 2 - x[i]**2

plt.plot(x,y,'s-')
plt.xlabel('x'), plt.ylabel('y')
plt.show()
```

```python
# what if you want more spacing?

x = np.linspace(-3,4,14)
y = 2 + np.sqrt(abs(x))

plt.plot(x,y,'s-')
plt.show()
```

```python
e = range(-1,4)
x = np.linspace(-4,4,300)

for i in e:
  y = x**i
  plt.plot(x,y,label='$y=x^{%s}$'%i,linewidth=4)

plt.legend()
plt.ylim([-20,20])
plt.xlim([x[0],x[-1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
