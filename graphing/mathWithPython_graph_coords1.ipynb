```python
# import libraries
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5

# basic plotting a red dot
plt.plot(x,y,'ro')

# set axis limits
plt.axis('square') # order matters
plt.axis([-6,6,-6,6])
plt.grid()

plt.show()
```

```python
# a set of coordinates

x = [-4,2,5,6,2,-5]
y = [5,2,10,-5,4,0]

for i in range(0,len(x)):
  plt.plot(x[i],y[i],'o',label='point %s'%i)


plt.legend()
plt.axis('square')
plt.grid()
plt.show()
```

```python
# getting information from axes

plt.plot(4,3,'rs')

# get an object for the current axis
axis = plt.gca()
ylim = axis.get_ylim()
print(ylim)

# now change only the upper y-axis limit
axis.set_ylim([ ylim[0],6 ])

plt.xlabel('X axis')
plt.ylabel('F(x)')

plt.show()
```
